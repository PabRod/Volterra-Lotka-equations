```python
from scipy.integrate import odeint #Used for solving the equation
from pylab import * #For using plot
%matplotlib inline
```

###Volterra-Lotka equations

The Volterra-Lotka equations are the basic mathematical model for predator-prey interactions. They look like:

$$x'(t) = a \cdot x(t) - b \cdot x(t) \cdot y(t)$$
$$y'(t) = -c \cdot y(t) + d \cdot x(t) \cdot y(t)$$

with: 

$$x(0) = x_0$$
$$y(0) = y_0$$

```python
a = 1
b = 0.2
c = 1
d = 1

f = lambda x, t: [a*x[0] - b*x[0]*x[1], -c*x[1] + d*x[0]*x[1]]

x_0 = [2, 3]
```

###Problem solving

First of all, find the equilibrium solutions:

```python
x_eq, y_eq = c/d, a/b
print(x_eq)
print(y_eq)
```

Now solve it for a range in $t$:

```python
t = linspace(0, 4*pi, 250)

x = odeint(f, x_0, t)
```

Plot the results:

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10,5))

axes[0].plot(t, x[:,0], 'r')
axes[0].plot(t, x[:,1], 'b')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Population')
axes[0].set_title('Solutions')
axes[0].legend(('Prey','Predator'))

axes[1].plot(x[:,0], x[:,1], 'k')
axes[1].set_xlabel('Prey')
axes[1].set_ylabel('Predator')
axes[1].set_title('Phase map')
    
fig.tight_layout()
```
